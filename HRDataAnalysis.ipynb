```python
plt.figure(figsize=(10,6))
sns.countplot(data=data, x='JobRole')
plt.xticks(rotation=45, ha='right')  
plt.grid(axis='y', linestyle='--', alpha=0.7)  
```

```python
ax = sns.countplot(data=data, x='Gender', palette="mako", edgecolor='black')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'])['Department'].value_counts().sort_values(ascending=False)
```

```python
data.groupby(['Gender'])['EducationField'].value_counts().sort_values(ascending=False)
```

```python
data.groupby(['Gender'])['Attrition'].value_counts().sort_values(ascending=False)
```

```python
grouped_data = data.groupby(['Gender', 'Attrition']).size().reset_index(name='Counts')

plt.figure(figsize=(12, 6))
attritionGender = sns.barplot(x='Gender', y='Counts', hue='Attrition', data=grouped_data, palette="autumn", edgecolor='black')

for container in attritionGender.containers:
    attritionGender.bar_label(container)

plt.xticks(rotation=0, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
